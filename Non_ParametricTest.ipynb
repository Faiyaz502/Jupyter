#Non-Parametric Tests
A non-parametric test is a statistical test that does not assume the data follows a specific distribution (like the normal distribution). These tests are useful when:

Data is ordinal (ranked),
Sample size is small,
Data is skewed or not normally distributed,
There are outliers

They are sometimes called distribution-free tests.

##Examples of non-parametric tests:

1. Mann–Whitney U Test

2. Wilcoxon Signed-Rank Test

3. Kruskal–Wallis Test

4. Spearman Rank Correlation

5. Chi-Square Test


| Purpose                        | Parametric Test     | Non-Parametric Alternative |
| ------------------------------ | ------------------- | -------------------------- |
| Compare two independent groups | Independent t-test  | Mann–Whitney U test        |
| Compare two related groups     | Paired t-test       | Wilcoxon Signed-Rank test  |
| Compare more than two groups   | ANOVA               | Kruskal–Wallis test        |
| Correlation                    | Pearson correlation | Spearman rank correlation  |
| Categorical association        | —                   | Chi-square test            |


```python
import numpy as np
from scipy.stats import mannwhitneyu
from scipy.stats import wilcoxon
from scipy.stats import kruskal
from scipy.stats import friedmanchisquare
from scipy.stats import spearmanr
from scipy.stats import chi2_contingency
```

# ============================================================
#                    2. MANN-WHITNEY U TEST
# ============================================================

##Definition:
The Mann-Whitney U Test compares two independent groups.

Used instead of:
Independent t-test

Example:
Comparing marks of two different classes.


```python
class_A = [12, 15, 14, 10, 18]
class_B = [22, 25, 19, 24, 20]

state,p = mannwhitneyu(class_A,class_B)

print("State :",state)
print("P : ",p)

if p < 0.05:
    print("Significant difference between groups")
else:
    print("No significant difference")
```

# ============================================================
# 3. WILCOXON SIGNED-RANK TEST
# ============================================================

##Definition:
The Wilcoxon Signed-Rank Test compares two related samples.

Used instead of:
Paired t-test

Example:
Before and after treatment scores.


```python
before = [10, 12, 15, 14, 13]
after  = [12, 15, 18, 17, 16]

stat , p = wilcoxon(before,after)

print("State :",stat)
print("P : ",p)

if p > 0.5 :
  print("No significant difference between groups")
else:
  print("Significant difference between groups")
```

# ============================================================
# 4. KRUSKAL-WALLIS TEST
# ============================================================

##Definition:
The Kruskal-Wallis Test compares more than two independent groups.

Used instead of:
One-Way ANOVA

Example:
Comparing scores of three classes.


```python

group1 = [12, 14, 16, 18]
group2 = [22, 24, 26, 28]
group3 = [30, 32, 34, 36]

stat, p = kruskal(group1, group2, group3)

print("\nKruskal-Wallis Test")
print("Statistic =", stat)
print("P-value =", p)

if p < 0.05:
    print("At least one group is different")
else:
    print("No significant difference")
```

# ============================================================
# 5. FRIEDMAN TEST
# ============================================================

##Definition:
The Friedman Test compares more than two related groups.

Used instead of:
Repeated Measures ANOVA

Example:
Same students tested under three methods.



```python
method1 = [7, 8, 6, 9]
method2 = [8, 9, 7, 10]
method3 = [9, 10, 8, 11]

stat, p = friedmanchisquare(method1, method2, method3)

print("\nFriedman Test")
print("Statistic =", stat)
print("P-value =", p)

if p < 0.05:
    print("Significant difference among methods")
else:
    print("No significant difference")
```

# ============================================================
# 6. SPEARMAN RANK CORRELATION
# ============================================================


##Definition:
Spearman Rank Correlation measures relationship
between two ranked variables.

Used instead of:
Pearson Correlation

Example:
Relationship between study rank and exam rank.



```python
study_rank = [1, 2, 3, 4, 5]
exam_rank  = [2, 1, 4, 3, 5]

corr, p = spearmanr(study_rank, exam_rank)

print("\nSpearman Rank Correlation")
print("Correlation =", corr)
print("P-value =", p)

if p < 0.05:
    print("Significant correlation")
else:
    print("No significant correlation")
```

# ============================================================
# 7. CHI-SQUARE TEST
# ============================================================


##Definition:
The Chi-Square Test checks association
between categorical variables.

Example:
Relationship between gender and product preference.


```python

data = [[30, 20],
        [15, 35]]

stat, p, dof, expected = chi2_contingency(data)

print("\nChi-Square Test")
print("Statistic =", stat)
print("P-value =", p)
print("Degrees of Freedom =", dof)
print("Expected Frequencies =", expected)

if p < 0.05:
    print("Variables are associated")
else:
    print("Variables are independent")
```
