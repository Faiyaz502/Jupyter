##A company claims the average battery life is 10 hours

sample size = 36

Test at 5% significance level.

```python
import math
```

```python
import math

n = 36
population_std_sigma = 1.2
x_bar = 9.5
mu = 10

z = (x_bar - mu)/(population_std_sigma/math.sqrt(n))
print("Z value:", z)
```

z value is less then -1.96 so Reject null hypotheses

Answer

There is sufficient evidence that the average battery life is different from 10 hours.

80 passed out of 100

70 passed out of 100

Test whether pass percentages differ significantly.

```python
from statsmodels.stats.proportion import proportions_ztest

count = [80,70]
nobs = [100,100]


z_stat,p_val = proportions_ztest(count,nobs)

print("Z value :",z_stat,"P-Value :",p_value)


alpha = 0.05

if p_val > alpha :
  print("Fail to Reject Null Hypothesis")
else:
  print("Reject Null Hypothesis")
```

#1. One Sample T-Test

##Used to compare a sample mean with a population mean when population standard deviation is unknown.

Example

A teacher claims average marks = 70.

Sample marks:

[72, 75, 68, 70, 69]

Test at 5% significance level.

```python
from scipy.stats import ttest_1samp


claim_mean = 70
sample = [72,75,68,70,69]

t_stat,p_value = ttest_1samp(sample,claim_mean)

print("T value :",t_stat,"P-Value :",p_value)

alpha = 0.05

if alpha < p_value :
  print("Fail to Reject Null Hypothesis")
else:
  print("Reject Null Hypothesis")
```

#2.Independent (Two-Sample) T-Test
##Definition

##Used to compare means of two independent groups.

Salary comparison:

Company A: [45, 50, 55, 60]
Company B: [40, 42, 48, 46]
its same or not
	​


```python

from scipy.stats import ttest_ind


group1 = [45, 50, 55, 60]
group2 = [40, 42, 48, 46]

t_stat,p_value = ttest_ind(group1,group2)

print("T value :",t_stat,"P-Value :",p_value)

alpha = 0.05

if alpha < p_value :
  print("Fail to Reject Null Hypothesis")
  print("Same")
else:
  print("Reject Null Hypothesis")
```

#3. Paired T-Test
##Definition

Used when same group is measured before and after treatment.

##Weight before and after diet:

##Before: [80, 85, 78]
##After: [75, 82, 76]




```python
from scipy.stats import ttest_rel

group1 = [80,85,78]
group2 = [75,82,76]

t_stat,p_value = ttest_rel(group1,group2)

print("T value :",t_stat,"P-Value :",p_value)

alpha = 0.05

if p_value < alpha :
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

#One-Way ANOVA
## Definition

##Used to compare 3 or more group means.

Example

Exam scores:

A = [70, 75, 80]
B = [65, 68, 70]
C = [90, 88, 92]

```python
from scipy.stats import f_oneway

A = [70, 75, 80]
B = [65, 68, 70]
C = [90, 88, 92]

f_stat, p_value = f_oneway(A, B, C)

print("F-statistic:", f_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

#Pearson Correlation Test
##Definition

##Measures linear relationship between two numerical variables.

Example

Hours studied vs marks:

hours = [1,2,3,4,5]
marks = [50,55,65,70,80]

```python
from scipy.stats import pearsonr

hours = [1,2,3,4,5]
marks = [50,55,65,70,80]

corr, p_value = pearsonr(hours, marks)

print("Correlation:", corr)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
